```python
# Here's an example using z-score method to detect and remove outliers in a specific column (replace 'column_name' with the actual column name):
from scipy.stats import zscore
```

```python
z_scores = zscore(df.select_dtypes(include=['float64', 'int64']))
threshold = 3
outlier_mask = (abs(z_scores) > threshold).any(axis=1)
df_cleaned = df[~outlier_mask]
# Now df_cleaned contains the DataFrame without outliers
```

```python
# Defining a function to calculate the IQR for each columns
def IQR(df,column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return print(IQR)
```

### 3.Data Visualization:

```python
# Histograms or boxplots for numerical variables
numerical_variables = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP']
for var in numerical_variables:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[var], kde=True, bins=20)
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Scatter plots or correlation heatmaps to explore relationships between pairs of variables

plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot
sns.pairplot(df)
plt.show()
```

```python
# Violin Plot is a method to visualize the distribution of numerical data
fig,axs=plt.subplots(7,2,figsize=(12,28))
sns.violinplot(ax=axs[0,0],data=df,y='LB')
sns.violinplot(ax=axs[0,1],data=df,y='AC')
sns.violinplot(ax=axs[1,0],data=df,y='FM')
sns.violinplot(ax=axs[1,1],data=df,y='UC')
sns.violinplot(ax=axs[2,0],data=df,y='DL')
sns.violinplot(ax=axs[2,1],data=df,y='DS')
sns.violinplot(ax=axs[3,0],data=df,y='DP')
sns.violinplot(ax=axs[3,1],data=df,y='ASTV')
sns.violinplot(ax=axs[4,0],data=df,y='MSTV')
sns.violinplot(ax=axs[4,1],data=df,y='ALTV')
sns.violinplot(ax=axs[5,0],data=df,y='MLTV')
sns.violinplot(ax=axs[5,1],data=df,y='Width')
sns.violinplot(ax=axs[6,0],data=df,y='Tendency')
sns.violinplot(ax=axs[6,1],data=df,y='NSP')
plt.show()
```
