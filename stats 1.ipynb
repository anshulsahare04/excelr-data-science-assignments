## Standardization of Numerical Variables

```python
df1.head()
```

```python
Stand=pd.DataFrame(columns=['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value'])    #Empty dataframe
Stand
```

```python
# Calculating the standardize of numerical columns(z score normilation)

Stand['Volume']=(df1['Volume']-df1['Volume'].mean())/df1['Volume'].std()

Stand['Avg Price']=(df1['Avg Price']-df1['Avg Price'].mean())/df1['Avg Price'].std()

Stand['Total Sales Value']=(df1['Total Sales Value']-df1['Total Sales Value'].mean())/df1['Total Sales Value'].std()

Stand['Discount Rate (%)']=(df1['Discount Rate (%)']-df1['Discount Rate (%)'].mean())/df1['Discount Rate (%)'].std()

Stand['Discount Amount']=(df1['Discount Amount']-df1['Discount Amount'].mean())/df1['Discount Amount'].std()

Stand['Net Sales Value']=(df1['Net Sales Value']-df1['Net Sales Value'].mean())/df1['Net Sales Value'].std()
```

## Conversion of Categorical Data into Dummy Variables

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
```

```python
df.head()
```

```python
df2
```

```python
df_encoding= pd.get_dummies(df2,columns=['Day','Date'],drop_first=True)
df_encoding
```

```python
df_encoding=pd.get_dummies(df2, columns=['SKU'],drop_first=True)
df_encoding
```

```python
# Creating variable of functions to make is easy for using multiple times.
one_hot = make_pipeline(OneHotEncoder())
ord_enc = make_pipeline(OrdinalEncoder())
```

```python
# Using one hot encoder 
transformers=ColumnTransformer([('one_hot',one_hot,['BU'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```

```python
transformers=ColumnTransformer([('one_hot',one_hot,['City'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```

```python
# Using ordinal encoder

transformers=ColumnTransformer([('ord_enc',ord_enc,['Brand'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```

```python
transformers=ColumnTransformer([('ord_enc',ord_enc,['Model'])])
df=pd.DataFrame(transformers.fit_transform(df2),columns=transformers.get_feature_names_out())
df.head()
```
